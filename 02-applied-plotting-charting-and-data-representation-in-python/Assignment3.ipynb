In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

n = 3650
df = pd.DataFrame([np.random.normal(32000,200000,n), 
                   np.random.normal(43000,100000,n), 
                   np.random.normal(43500,140000,n), 
                   np.random.normal(48000,70000,n)], 
                  index=[1992,1993,1994,1995])
df
```

```python
import ipympl
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
df.T.describe()
```

```python
import scipy.stats as st

# Confidence interval adapted from https://stackoverflow.com/a/34474255
def confidence_interval(array, ci=0.95):
    return st.t.interval(ci, len(array)-1, loc=0, scale=st.sem(array))

data = pd.DataFrame()
data['mean'] = df.mean(axis=1)
data['ci'] = df[list(range(n))].apply(lambda row: confidence_interval(row)[1], axis=1)

data
```

```python
# Color Helper from SO: https://stackoverflow.com/a/26109298
import matplotlib as mpl
from matplotlib import cm

class MplColorHelper:

  def __init__(self, cmap_name, start_val, stop_val):
    self.cmap_name = cmap_name
    self.cmap = plt.get_cmap(cmap_name)
    self.norm = mpl.colors.Normalize(vmin=start_val, vmax=stop_val)
    self.scalarMap = cm.ScalarMappable(norm=self.norm, cmap=self.cmap)

  def get_rgb(self, val):
    return self.scalarMap.to_rgba(val)

MplColorHelper('viridis', 50, 100).get_rgb(75)
```

```python
bar_width = 2
num_bars = 4
major_ticks = np.arange(0, num_bars*bar_width, bar_width)
minor_ticks = major_ticks + bar_width/2
xlim = [0, num_bars*bar_width]

color_map = 'RdBu' # This color map matches the colors in the example
color_helpers = [MplColorHelper(color_map, row['mean'] - row['ci'], row['mean'] + row['ci']) for _, row in data.iterrows()]

# Call this every time we need to re-draw the figure
def draw(selected):
    ax = plt.gca()

    # Bar chart
    ax.bar(major_ticks, data['mean'], bar_width, align='edge', yerr=data['ci'], capsize=10, color=[ch.get_rgb(selected) for ch in color_helpers])
    
    # Plot selected value as horizontal line
    ax.plot(xlim, [selected, selected], 'gray')

    # Make graph look nice(r)
    plt.xlim(xlim)
    plt.xticks(major_ticks, [""]*len(major_ticks))
    ax.set_xticks(minor_ticks, minor=True)
    ax.set_xticklabels(data.index, minor=True, fontsize='large')
    for line in ax.xaxis.get_minorticklines():
        line.set_visible(False)

    
# Call this when the user clicks somewhere. Update the selected value.
def onclick(event):
    selected = event.ydata
    
    plt.cla()
    draw(selected)

default = 40000
fig = plt.figure(figsize=(10,8))
fig.canvas.mpl_connect('button_press_event', onclick)
draw(default)
```
